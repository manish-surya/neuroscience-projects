# Neuron Membrane Potential Simulation

This Project simulates the membrane potential dynamics of a neuron using an RC circuit model.

Concepts demonstrated:
- Membrane potential
- RC circuit analogy
- Time constant
- Numerical integration (Euler method)

Python libraries:
- numpy
- plotly

### Import libraries

```python
import numpy as np
import plotly.graph_objects as go
```

### Neuron Model Class

```python
class NeuronMembraneModel:
    """
    A simple neuron membrane model using RC circuit dynamics.
    """

    def __init__(self,
                 resting_potential = -70.0,
                 resistance = 10.0,
                 capacitance = 1.0,
                 dt = 0.1,
                 simulation_time = 200):
        """
        Parameters
        -----------
        resting_potential: float
            Resting membrane potential - The electric potential difference across
            the cell membrane at rest, typically around -70 mV, with the inside of
            the neuron being more negative than the outside.
            (Unit - mV)

        resistance: float
            Membrane resistance - Membrane resistance (R) represents the ability
            of the cell membrange to prevent ion flow, largely determined by the
            number of open ion channels.
            High resistance means fewer channels are open, leading to larger voltage
            changes from a given current (v = IR)
            (Unit - MΩ)

        capacitance: float
            Membrane capacitance - ability of the cell membrane to store electric
            charge, typically measured in microfarads (μF)

        dt: float
            simulation timestep (ms)

        simulation_time: int
            total simulation duration (ms)

        """

        self.V_rest = resting_potential
        self.R = resistance
        self.C = capacitance
        self.dt = dt
        self.T = simulation_time

        # Time constant
        self.tau = self.R * self.C

        # Time array
        self.time = np.arange(0, self.T, self.dt)

        # Membrane Voltage Initialization
        self.voltage = np.zeros(len(self.time))
        self.voltage[0] = self.V_rest

    # Current Stimulus
    def external_current(self, t):
        """
        Defines injected current stimulus.
        Current is applied between 50 ms and 150 ms.
        """
        if 50 <= t <=150:
            return 1.5
        return 0.0

    # Simulation Engine
    def simulation(self):
        """
        Runs teh neuron membranesimulation using
        Euler numerical integration.
        """

        for i in range(1, len(self.time)):
            t = self.time[i]

            #current input
            I = self.external_current(t)

            #previous voltage
            V_prev = self.voltage[i-1]

            #differential equation
            dVdt = (-(V_prev - self.V_rest) + self.R * I) / self.tau

            # Euler update
            self.voltage[i] = V_prev + self.dt * dVdt

        return self.time, self.voltage
```

### Visualization

```python
def plot_voltage(time, voltage):

    fig = go.Figure()

    fig.add_trace(
        go.Scatter(
            x = time,
            y = voltage,
            mode = 'lines',
            name="Membrane Potential"
        )
    )

    fig.update_layout(
        title="Neuron Membrane Potential Simulation",
        xaxis_title = "Time (ms)",
        yaxis_title = "Membrane Voltage (mV)",
        template = "plotly_white"
    )

    fig.show()
```

### Main Execution

```python
def main():
    print("\n Running Neuron Membrane Simulation...\n")
    neuron = NeuronMembraneModel()
    time, voltage = neuron.simulation()
    plot_voltage(time, voltage)
    print("\n Simulation complete\n")


if __name__ == "__main__":
    main()
```
